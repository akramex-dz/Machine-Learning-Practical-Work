```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.metrics import accuracy_score
from ML_TP3_tools import *
```

***EXERCISE NUMBER 1***

```python
class0 = np.array([[1, 9], [5, 5], [1, 1]])
class1 = np.array([[8, 5], [13, 1], [13, 9]])

X = np.vstack((class0, class1))
y = np.array([0, 0, 0, 1, 1, 1])

print(X)
print(y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
ax = plt.gca()
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X, y)
```

```python
xx, yy = make_meshgrid(X, y, 0.01)
plot_decision_boundary(X, y, xx, yy, xx > 6.5, plot_mean=True)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T

Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 2], alpha=0.5,
           linestyles=['--', '-', '--'])

ax.scatter(clf.support_vectors_[:, 0],
           clf.support_vectors_[:, 1],
           s=200,
           linewidth=2,
           facecolors='none',
           edgecolors='blue')

plt.show()
```

***EXERCICE NUMBER 2***

```python
class0 = np.array([[2, 2], [3, 3], [4, 5], [5, 5], [4, 6], [3, 7], [4, 8], [5, 9], [6, 10]])
class1 = np.array([[6, 2], [7, 3], [8, 4], [9, 5], [8, 6], [7, 7], [7, 8], [7, 9], [8, 10]])

X = np.vstack((class0, class1))
y = np.array([0] * len(class0) + [1] * len(class1))

print(y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', marker='o')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Binary Dataset')
```

```python
clf = svm.SVC(kernel='linear', C=1e6)
clf.fit(X, y)

support_vectors = clf.support_vectors_

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', marker='o')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Binary Dataset')

plt.scatter(
    support_vectors[:, 0], 
    support_vectors[:, 1], 
    color='red', 
    marker='x', 
    s=50, 
    label='Support Vectors')

w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(X[:, 0].min(), X[:, 0].max())
yy = a * xx - (clf.intercept_[0]) / w[1]

plt.plot(xx, yy, 'k--', label='Decision Boundary')

plt.legend()
plt.show()

y_pred = clf.predict(X)
error_rate = 1 - accuracy_score(y, y_pred)
print(f"Error rate on training data: {error_rate:.4f}")
```

```python
xx, yy = make_meshgrid(X, y, 0.01)

clf = svm.SVC(kernel='poly', gamma=1000 ,C=1e1)
clf.fit(X, y)

Z = clf.predict(np.array([xx.ravel(), yy.ravel()]).T).reshape(xx.shape)
plot_decision_boundary(X, y, xx, yy, Z, plot_mean=True)
```
