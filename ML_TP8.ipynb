```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=500, n_features=2, centers=3, random_state=1)

fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
import numpy as np

k=3
clusters = {}
np.random.seed(23)
for i in range(k):
    center = 2 * ( 2 * np.random.random( (X.shape[1], ) ) - 1 )
    points = []
    cluster = {'center' : center,'points' : [] }
    clusters[i] = cluster

# print(clusters)

plt.scatter(X[:,0],X[:,1])
plt.grid(True)

for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0],center[1],marker = '*',c = 'red')
plt.show()
```

```python
import numpy as np

def distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))

def assign_clusters(X, clusters):
    for i in range(X.shape[0]):
        dist = []
        curr_x = X[i]
        for j in range(k):
            dis = distance(curr_x, clusters[j]['center'])
            dist.append(dis)
        curr_cluster = np.argmin(dist)
        clusters[curr_cluster]['points'].append(curr_x)
    return clusters
```

```python
assigned_clusters = assign_clusters(X, clusters)

plt.scatter(X[:,0],X[:,1])
plt.grid(True)

for i in assigned_clusters:
    center = assigned_clusters[i]['center']
    plt.scatter(center[0],center[1],marker = '*',c = 'red')
plt.show()
```

```python
def update_clusters(X, clusters):
    for i in range(k):
        points = np.array(clusters[i]['points'])
        if points.shape[0] > 0:
            new_center = points.mean(axis =0)
            clusters[i]['center'] = new_center
            clusters[i]['points'] = []
    return clusters
```

```python
def clusering(data, centroids):

    cluster_membership = assign_clusters(data, centroids)

    while True:
        centroids = update_clusters(data, cluster_membership)

        new_cluster_membership = assign_clusters(data, centroids)

        if (new_cluster_membership == cluster_membership).all():
            break  
        
        cluster_membership = new_cluster_membership
    return cluster_membership
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

iris = load_iris()
X = iris.data

kmeans = KMeans(n_clusters=5)
kmeans.fit(X)

labels = kmeans.labels_

print(labels)

import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Cluster Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = np.array(X)

wcss = []

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    
    kmeans.fit(data)
    
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Curve')
plt.show()
```
