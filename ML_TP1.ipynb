```python
positive_df = covid[covid['SARS-Cov-2 exam result'] =='positive']
negative_df = covid[covid['SARS-Cov-2 exam result'] =='negative']
```

```python
missing_rate =covid.isna().sum()/covid.shape[0]
```

```python
blood_columns = covid.columns[(missing_rate < 0.9) & (missing_rate >0.88)]
```

```python
viral_columns = covid.columns[(missing_rate < 0.88) & (missing_rate >0.75)]
```

```python
for col in blood_columns:
    plt.figure()
    sns.distplot(positive_df[col],label='positive')
    sns.distplot(negative_df[col], label='negative')
    plt.legend()
```

```python
sns.countplot(x='Patient age quantile', hue='SARS-Cov-2 exam result', data=covid)
```

```python
pd.crosstab(covid['SARS-Cov-2 exam result'], covid['Influenza A'])
```

```python
for col in viral_columns:
    plt.figure()
    sns.heatmap(pd.crosstab(covid['SARS-Cov-2 exam result'], covid[col]))
```

```python
# Data
covid_results = covid["SARS-Cov-2 exam result"].value_counts()

# Create bar chart
plt.bar(covid_results.index, covid_results, color =['green', 'red'], width = 0.4)

# Labeling the axes and title²plt.xlabel("SARS-Cov-2 exam result")
plt.ylabel("No. of cases")
plt.title("Positive vs Negative covid cases")

# Display the chart
plt.show()
```

```python
covid['SARS-Cov-2 exam result'] = covid['SARS-Cov-2 exam result'].replace('negative', '0').astype(float)
covid['SARS-Cov-2 exam result'] = covid['SARS-Cov-2 exam result'].replace('positive', '1').astype(float)
covid['SARS-Cov-2 exam result'] = covid['SARS-Cov-2 exam result'].astype(float)

corr_matrix = covid.corr()
print(corr_matrix["SARS-Cov-2 exam result"].sort_values(ascending=False))
```
